```python
import pandas as pd
df = pd.read_csv("D:\\Elevate Data Analyst\\netflix_titles.csv\\netflix_titles.csv")  # adjust path as needed


print(df.info())
print(df.head())
```

```python
# Check for nulls
print(df.isnull().sum())
```

```python
# Fill or drop missing values
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Not Specified', inplace=True)
df['country'].fillna('NA',inplace=True)
df['rating'].fillna('NA',inplace=True)
df['duration'].fillna('NA',inplace=True)
df.dropna(subset=['date_added'], inplace=True)
```

```python
print(df.isnull().sum())
```

```python
#remove duplicates
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
# standardizing text
df['country'] = df['country'].str.strip().str.title()
```

```python
df.head()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'].str.strip())
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
df.head()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.head()
```

```python
print(df.dtypes)
# Check if release_year is integer
df['release_year'] = df['release_year'].astype(int)
```

```python
df.to_csv("D:\\Elevate Data Analyst\\netflix_cleaned.csv", index=False)
```
